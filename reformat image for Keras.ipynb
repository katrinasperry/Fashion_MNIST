```python
import numpy as np
import pandas as pd
from PIL import Image
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
# load JPEG image
jpgimage= Image.open('realreal_test_img.jpg')
type(jpgimage)

#print(jpgimage.bits, jpgimage.size, jpgimage.format)
```

```python
# display original image
plt.imshow(jpgimage)
plt.show()
```

```python
# convert to numpy array
image = np.asarray(jpgimage)
type(image)
```

```python
# read image into cv2 format, resize to 28 x 28
image=cv2.resize(image,(28,28))

# use swapaxes to convert image to Keras' format
image_convert=np.swapaxes(np.swapaxes(image, 1, 2), 0, 1)

type(image)
```

```python
# show that it worked
print(image_convert.shape)
plt.imshow(image_convert[1,:,:], cmap=plt.cm.gray)
plt.show()
```
